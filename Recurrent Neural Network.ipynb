# Сеть встречного распространения

Сеть встречного распространения является одним из типов нейронных сетей, разработанных для анализа и моделирования последовательных данных, таких как текст, звук, временные ряды и т.д. Отличительной особенностью RNN является возможность использования информации из предыдущих шагов (прошлого контекста) в текущем шаге обработки, что делает ее мощным инструментом для анализа последовательностей.

Нейронная сеть встречного распространения в своей архитектуры объединяет преимущества обобщения сети Т. Кохонена и простоту обучения вы­ходной звезды Д. Гроссберга, из-за чего сеть получила свойства, которых нет ни у одной сетей по отдельности. Она принадлежит к классу сетей, обучающихся без учителя.

*В сети Кохонена происходит решение задачи нахождения кластеров в пространстве входных образов. Сеть Кохонена обучается без учителя на основе самоорганизации. В течении обучения вектора весов нейронов стремятся к центрам кластеров – групп векторов обучающей выборки. После обучения сеть сопоставляет предъявляемый образ к одному из кластеров, то есть к одному из выходов. В общем случае задача кластеризации представляется следующим образом:*
- *есть объекты, характеризуемые вектором  параметров ![](image003.gif), имеющих ![](image004.gif) компонент ![](image006.gif);*
- *есть введенное множество классов ![](image008.gif) в пространстве классов ![](image010.gif) (обычно ![](image012.gif), в случае ![](image014.gif), задача сводится к тривиальной);*
